```python
%pylab inline
from scipy.optimize import curve_fit

def fun(x, a, b, c, d):
    return a*x**3+b*x**2+c*x+d

x= np.array([3,4,5,7,8,9,11,12])
y= np.array([1.6,3.6,4.4,3.4,2.2,2.8,3.8,4.6])

cons, npi = curve_fit(fun,x,y)

print(cons)
```

```python
t=linspace(x[0],x[-1])
plt.plot(x,y,'*',t,fun(t,*cons))
plt.xlabel('x')
plt.ylabel('y')
```
