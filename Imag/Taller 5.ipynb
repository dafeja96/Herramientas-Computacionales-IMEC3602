En el presente informe se expone el proceso llevado a cabo para realizar el procesamiento de una imagen con el fin de poder estimar los tamaños de granos de una micrografía. Como primer paso se importan todas las librerías necesarias:

```python
%pylab inline
import scipy
from scipy import misc
from scipy import ndimage
from skimage import data, color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte
import cv2
```

##### Ahora bien, se va a hacer el análisis de granulometría.

```python
img = cv2.imread('FeNod_50X_L.tif')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
kernel = np.ones((4,4),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
#closing = cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel, iterations = 2)

figure(figsize=(15,10))
plt.subplot(2,2,1)
pylab.imshow(img)
plt.subplot(2,2,2)
pylab.imshow(thresh,'gray')
plt.subplot(2,2,3)
pylab.imshow(opening,'gray')
```

En estas primeras tres imágenes se observa primero la imagen original, en la segunda se observa la imagen bajo un proceso de binarización y en la tercera bajo la creación de un kernel se utilizó cv2.morphologyEx lo cual permitió eliminar algo de ruido en la imagen.

```python
bgt = cv2.dilate(opening,None,iterations = 2)
pylab.imshow(bgt,'gray')
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,3)
ret, fg = cv2.threshold(dist_transform,0.1*dist_transform.max(),255,0)
figure()
pylab.imshow(dist_transform,'gray')
figure()
pylab.imshow(fg,'gray')
```

```python
fg = np.uint8(fg)
unknown = cv2.subtract(bgt,fg)

ret, markers = cv2.connectedComponents(fg)

markers = markers+1

markers[unknown==255] = 0

markers = cv2.watershed(img,markers)
edges=markers==-1
img[edges] = [255,0,0]

figure()
pylab.imshow(unknown,'gray')
figure()
pylab.imshow(img,'gray')
```

```python
hough_res = hough_circle(edges,np.arange(50,80,1))
accums, cx, cy, radii = hough_circle_peaks(hough_res, np.arange(50,80,1))
```

```python
img = cv2.imread('FeNod_50X_L.tif')
crop_img = img[1750:2000,2250:2820, :]
figure()
plt.imshow(crop_img)
n=4000
for centery,centerx,radios in zip(cy[0:n], cx[0:n], radii[0:n]):
    circy, circx = circle_perimeter(centery, centerx, radios)
    if (np.max(circy)<img.shape[0] and np.max(circx) <img.shape[1]):
        img[circy, circx] = (250,0,0)
figure()
pylab.imshow(img,'gray')
```
